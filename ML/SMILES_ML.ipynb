```python
import pandas as pd
import re
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit import DataStructs
import numpy as np
```

```python
path='../data/CycPeptMPDB_Peptide_All.csv'
```

```python
df = pd.read_csv(path)
df=df.drop_duplicates(subset=["Structurally_Unique_ID"], keep="first")
df= df.sample(frac=1).reset_index(drop=True)
```

```python
def smiles2fingerprint(smi):
    mol=Chem.MolFromSmiles(smi)
    fingerprint = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
    fp_array = np.zeros((1,), dtype=np.int8)
    DataStructs.ConvertToNumpyArray(fingerprint, fp_array)
    return fp_array
                           
df['fingerprint']=df['SMILES'].apply(smiles2fingerprint)
labels = df['Permeability'].values
all_smi_fg=df['fingerprint'].values
all_smi_fg = np.vstack(all_smi_fg)
```

# 随机森林

```python
import pandas as pd
from sklearn.model_selection import train_test_split

def split_dataframe(df, train_ratio=0.6, test_ratio=0.2, val_ratio=0.2, random_seed=42):
    """
    Split a DataFrame into training, testing, and validation sets.

    Parameters:
    - df: The input DataFrame
    - train_ratio: The ratio of the training set (default: 0.6)
    - test_ratio: The ratio of the test set (default: 0.2)
    - val_ratio: The ratio of the validation set (default: 0.2)
    - random_seed: Random seed for reproducibility

    Returns:
    - train, test, val: Training, testing, and validation DataFrames
    """
    
    # Check if ratios sum up to 1
    if train_ratio + test_ratio + val_ratio != 1.0:
        raise ValueError("Ratios must sum up to 1.")
    
    # Split into train and temp dataframes
    train, temp = train_test_split(df, test_size=test_ratio + val_ratio, random_state=random_seed)
    
    # Calculate test to temp ratio
    test_to_temp_ratio = test_ratio / (test_ratio + val_ratio)
    
    # Split temp into test and validation dataframes
    test, val = train_test_split(temp, test_size=1-test_to_temp_ratio, random_state=random_seed)

    return train, test, val

# 使用示例

df = pd.read_csv('/home/s2136015/Code/wzx_utils/data/cyy_use_CycPeptMPDB_Peptide_All.csv')
df=df.drop_duplicates(subset=["Structurally_Unique_ID"], keep="first")
df= df.sample(frac=1).reset_index(drop=True)

train_df, test_df, val_df = split_dataframe(df)

print(f"Training set size: {len(train_df)}")
print(f"Test set size: {len(test_df)}")
print(f"Validation set size: {len(val_df)}")

train_df.to_csv('../data/train.csv',index=False)
test_df.to_csv('../data/test.csv',index=False)
val_df.to_csv('../data/val.csv',index=False)
```
